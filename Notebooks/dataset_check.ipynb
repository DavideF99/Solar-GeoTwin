```python
import os 
from pathlib import Path
import sys
from collections import Counter

# 1. Find the project root (one level up from /notebooks)
notebook_path = Path(os.getcwd())
PROJECT_ROOT = notebook_path.parent

# 2. Add root to sys.path so we can import our modules
sys.path.append(str(PROJECT_ROOT))

# 3. Define data paths relative to the root
IMAGE_DIR = PROJECT_ROOT / "data" / "images"
MASK_DIR = PROJECT_ROOT / "data" / "masks"
MODEL_PATH = PROJECT_ROOT / "models" / "solar_unet_v1.pth"

images = os.listdir(IMAGE_DIR)
masks = os.listdir(MASK_DIR)

print(f"Project Root identified as: {PROJECT_ROOT}")
```

```python
# Extract the area name (everything before '_patch_')
areas = [f.split('_patch_')[0] for f in images if f.endswith('.npy')]

stats = Counter(areas)

print("--- Global Dataset Distribution ---")
for area, count in stats.items():
    print(f"📍 {area.capitalize()}: {count} patches")
print(f"Total: {sum(stats.values())}")
```

```python
print(f"Total Images: {len(images)}")
print(f"Total Masks: {len(masks)}")

if len(images) != len(masks):
    print("⚠️ WARNING: Mismatch detected. Check for partial downloads.")
else:
    print("✅ Dataset is synchronized and ready for training.")
```
